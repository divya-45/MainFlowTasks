```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/kaggle/input/householdtask3-csv/householdtask3.csv')
```

```python
display(data.head(6))
```

```python
#scatter plot with year against own
plt.scatter(data['age'],data['size'])
#adding title to the plot
plt.title("scatter plot")
#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')
#adding the legends
plt.show()
```

```python
#line chart with year against own
plt.plot(data['year'])
plt.plot(data['own'])
#adding title to plot
plt.title('line chart')
#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')
#adding the legends
plt.show()
     
```

```python
#bar chart
plt.bar(data['prop_hhs'],data['own_prop'])
plt.title("bar chart")
#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')
#adding the legends
plt.show()
```

```python
#histogram
plt.hist(data['eqv_income'])
plt.title("histogram")
#adding the legends
plt.show()
```
